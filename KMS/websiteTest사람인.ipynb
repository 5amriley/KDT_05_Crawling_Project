```python
company = [x.split('|')[0].split('2024')[0].strip('(주) ').strip("(유") for x in result.keys()]

test = result[list(result.keys())[0]][0]


topheader_label = ['회사명', '회사의종류', '리뷰어분야', '직원여부', '지역', '날짜' ,'제목', '좋은점', '나쁜점', '희망사항(경영진에게)', '성장', '추천',
                   '승진기회가능성점수', '복지급여점수', '업무와 삶의 균형 점수', '사내문화점수', '경영진점수']
table = []
topheader_goodValue = [
    ['디자인', '영업/제휴', '전문직', '서비스/고객지원', '마케팅/시장조사', '개발', '데이터', '인사/총무', '기획/경영', '미디어/홍보', '생산관리/품질관리', '금융/재무', '유통/무역', '법률/법무', '교육', '엔지니어링', '생산/제조', '기타', '연구개발', '특수계층/공공', '의약'],
    ['현직원', '전직원'],
    ['서울', '기타', '인천', '해외', '부산', '경기', '대전', '경북', '대구', '경남', '울산', '광주', '충남', '제주', '전북', '세종']
]
for idx, companyTotalReview in enumerate(result.values()):
    # print(company[idx])
    for totalreviewidx, test in enumerate(companyTotalReview):
        souptest = BeautifulSoup(test,'lxml')
        
        reviews = souptest.select('div.content_wrap')
        for reviewidx, review in enumerate(reviews):
            if len(review.select('div.cont_discontinu')) == 0:
                
                # print(totalreviewidx ,reviewidx)
                topheader = review.select_one('div.content_top_ty2')
                topheader_value = []
                topheader_value.append(company[idx])
                # print(company[idx])
                topheader_value.append(souptest.select_one('div.info span').get_text().strip())
                topheader_result = topheader.select('span.txt1')
                topheader_result_tmp = [None for x in range(4)]
                for topheader_onesoup in topheader_result:
                    for goodvalue_idx, acceptlist in enumerate(topheader_goodValue):
                        valuetmp = topheader_onesoup.get_text().strip()
                        if valuetmp in acceptlist:
                            topheader_result_tmp[goodvalue_idx] = valuetmp
                        elif dateCheck.match(valuetmp):
                            topheader_result_tmp[3] = valuetmp

                topheader_value.extend(topheader_result_tmp)
                title = review.select_one('h2.us_label').get_text().split('\"')[1]
                good = review.select_one('dt.merit').find_next_sibling('dd').get_text().strip('\n ')
                bad = review.select_one('dt.disadvantages').find_next_sibling('dd').get_text().strip('\n ')
                wish = review.select_one('dl.tc_list').select('dd.df1')[-1].get_text().strip('\n ')
                sowhat = review.select_one('p.etc_box').get_text().strip('\n ').replace('\n', '')
                recommand = (review.select_one('p.txt.recommend.etc_box').get_text
                             ().strip('\n '))
                topheader_value.extend([title, good, bad, wish, sowhat, recommand])
                
                for score in review.select('div.bl_score'):
                    initExtract = score.attrs['style']
                    num = onlynum.search(initExtract).group().strip('%')
                    topheader_value.append(num)
                table.append(topheader_value)
            
    # if idx == 1:
    #     break

tmpdf = pd.DataFrame(table, columns = topheader_label)


tmpdf
    
    
```
